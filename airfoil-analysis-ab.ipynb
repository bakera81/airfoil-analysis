```python
import os
import scipy.io
import pandas as pd
import pprint
import matplotlib.pyplot as plt
```

```python
mat = scipy.io.loadmat('Re_3M_Clean.mat')
```

```python
mat.keys()
```

```python
cpc_pressure = mat['Cpc_pressure']
```

```python
cpc_df = pd.DataFrame(cpc_pressure)
```

```python
aoa = mat['alphaC']
```

```python
# making a list comprehension (instead of for)
aoa = [str(int(x[0])) + '_degrees' for x in aoa]
```

```python
# Create names for the columns
col_names = ["port_" + str(x) for x in range(39)]
col_num = [x for x in range(39)]
```

```python
# set the indexes (set up table)
cpc_df.index = aoa
cpc_df.columns = col_names

cpc_df
```

```python
cpc_stacked = cpc_df.stack()
cpc_stacked_df = cpc_stacked.to_frame()
```

```python
cpc_stacked_df.columns = ['pressure']
# cpc_stacked_df.reset_index
cpc_stacked_df.index.names = ['degree', 'port']
# cpc_stacked_df.reset_index(level=['port'])
# cpc_stacked_df.index.name
# cpc_stacked_df.loc['degree']
cpc_stacked_df
```

```python
cpc_stacked_df.reset_index(level='port', inplace=True)
cpc_stacked_df
```

```python
cpc_stacked_df.loc['0_degrees']
```

```python
numerify_port = lambda x: int(x.port[-2:])
test = cpc_stacked_df.assign(port_2=numerify_port)
test
```
